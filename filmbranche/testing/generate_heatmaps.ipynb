## Heatmap for Actors/Equipment/Location Tests

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Create the data
data = {
    "Prompt 1 Analyse aller Anforderungen": ["NOTOK", "OK", "NOTOK", "OK", "NOTOK", "NOTOK", "NOTOK", "OK", "NOTOK", "OK"],
    "Prompt 2 Analyse aller Anforderungen": ["NOTOK", "NOTOK", "OK", "NOTOK", "NOTOK", "OK", "OK", "NOTOK", "OK", "OK"],
    "Prompt 3 Analyse aller Anforderungen": ["NOTOK", "OK", "OK", "OK", "OK", "OK", "OK", "OK", "OK", "OK"]
}

df = pd.DataFrame(data).T
df.columns = range(1, 11)  # Set the columns to be the experiment numbers

# Plot the heatmap with white lines
plt.figure(figsize=(12, 6))
sns.heatmap(df.replace({"OK": 1, "NOTOK": 0}), annot=df, fmt="", cmap="RdYlGn", cbar=False, linewidths=.5, linecolor='white')
plt.title("Experiment Prompt Recognition")
plt.xlabel("Experiment Number")
plt.ylabel("Prompt")
plt.show()
```

## Heatmap for Actors/Equipment Tests

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Create the data 
data = {
    "Prompt 1 Analyse der Ressourcen": ["NOT OK", "NOT OK", "OK", "OK", "NOT OK", "OK", "NOT OK", "OK", "OK", "OK"],
    "Prompt 2 Analyse der Ressourcen": ["OK", "OK", "OK", "OK", "OK", "OK", "OK", "OK", "OK", "OK"]
}

# Convert the data to a DataFrame and transpose it
df = pd.DataFrame(data).T
df.columns = range(1, 11)  
# Plot the heatmap with white lines
plt.figure(figsize=(12, 4))
sns.heatmap(df.replace({"OK": 1, "NOT OK": 0}), annot=df, fmt="", cmap="RdYlGn", cbar=False, linewidths=.5, linecolor='white')
plt.title("Experiment Prompt Recognition")
plt.xlabel("Experiment Number")
plt.ylabel("Prompt")
plt.show()
```
